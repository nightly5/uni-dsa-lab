```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading data
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/mtcars.csv')
print(df.shape)
```

```python
print(df.columns)
print(df.columns[1:])
```

```python
data = [df[col] for col in df.columns[1:]]
print(data[0].head()) # just for verification
```

Boxplot

```python
plt.boxplot(data, labels=df.columns[1:])
plt.show()
```

Boxplot with each feature in some detail

```python
subplots = [plt.subplot(ax) for ax in (221, 222, 212)]
cols = (
    ('mpg', 'wt', 'qsec', 'carb'),
    ('disp', 'hp'),
    ('cyl', 'drat', *df.columns[8:-1]),
)
data = [[df[col] for col in colnames] for colnames in cols]
for subplot, d, labels in zip(subplots, data, cols):
  subplot.boxplot(d, labels=labels)
plt.show()
```

Scatterplot

```python
for feature in df.columns[2:]:
  plt.scatter(df.mpg, df[feature])
  plt.text(min(df.mpg), max(df[feature]), feature)
plt.show()
```

Scatterplot for each feature

```python
fig = plt.figure(figsize=(20, 35))
subplots = [
  fig.add_subplot(len(df.columns[2:]), 5, i+1)
    for i in range(len(df.columns[2:]))
]
for feature, subplot in zip(df.columns[2:], subplots):
  subplot.scatter(df.mpg, df[feature])
  subplot.set_title(f'{feature} vs mpg')
plt.show()
```

Line plot between mpg and each feature

```python
fig = plt.figure(figsize=(7,10))
subplots = [fig.add_subplot(ax) for ax in (211, 212)]
cols = (
    ('disp', 'hp'),
    (*df.columns[1:3], *df.columns[5:]),
)
data = [[df[col] for col in colnames] for colnames in cols]
for subplot, d, labels in zip(subplots, data, cols):
  for coldata, label in zip(d, labels):
    subplot.plot(coldata)
  subplot.legend(labels, bbox_to_anchor=(1.2, 1))
plt.show()
```

Misc: matrices

```python
import numpy as np
```

```python
x = np.matrix('1 2 3; 4 5 6')
y = np.matrix([[1, 2, 3], [4, 5, 6]])
sum = x + y
xT = x.T
yT = y.transpose()
dot = np.dot(x, yT)
print(f'x: {x}')
print(f'y: {y}')
print(f'sum: {sum}')
print(f'xT: {xT}')
print(f'yT: {yT}')
print(f'dot: {dot}')
print(f'dot other way: {x.dot(yT)}')
print(f'x: {x}')
print(f'diff: {x - y}')
```
