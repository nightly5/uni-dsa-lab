## Scatterplots and histograms

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [5, 7, 10, 12, 20]
y = [10, 20, 30, 40, 50]
y1 = [2.9 + 1.79 * xi for xi in x]

plt.scatter(x, y)
plt.scatter(x, y1)
plt.show()
```

```python
v = [
    10, 15, 38, 33, 44, 48, 50, 58, 59, 58, 60, 65,
    58, 30, 74, 70, 74, 75, 79, 88, 98, 98, 58
]
plt.hist(v)
plt.show()
```
