```python
import numpy as np
import pandas as pd
from io import BytesIO
from google.colab import files
from pandas.errors import EmptyDataError
```

```python
uploaded = files.upload()
f = None
try:
  f = BytesIO(uploaded.get('mtcars.csv'))
  df = pd.read_csv(f)
except EmptyDataError:
  print('Make sure the file name is mtcars.csv')
```

```python
print(df.shape)
df.head()
```

```python
x = df.iloc[:, 2:]
y = df.mpg
print(x.shape, tuple(x.columns), y.shape, sep="\n")
```

```python
from sklearn.model_selection import train_test_split

split_data = train_test_split(
    x, y, test_size=0.2, random_state=0
)

for data in split_data:
  print(data.shape)

x_train, x_test, y_train, y_test = split_data
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x_train, y_train)
r_sq = model.score(x, y)
bcap = np.insert(model.coef_, 0, model.intercept_, 0)
y_test_cap = model.predict(x_test)
print(
    f"Test data x\n{x_test}\n"
    f"Model parameters beta^\n{bcap}\n"
    f"Predicted response y^\n{y_test_cap}\n"
    f"Coefficient of determination: {r_sq}"
)
error = y_test - y_test_cap
sse = sum(error ** 2)
rmse = (sse / len(y_test)) ** 0.5
print(f"Sum square error: {sse}\nRoot mean square error: {rmse}")
```
